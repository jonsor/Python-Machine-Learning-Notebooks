# Encoding and Dimensionality Reduction

## One-hot Encoding

So far, weâ€™ve assumed that our data comes in as a two-dimensional array of floating-point  numbers, where each column  is a `continuous feature` that describes the datapoints.

But what if the data consist of `categorical features`? Categorical features are usually not numeric, and more often than not text. Examples of categorical features are the brand of a product, the color of a product, orthe department (books, clothing, hardware) it is sold in.

To use this kind of data we will need to encode it into a numeric represenation the machine learning algorithms can understand.

`One-hot encoding` is the technique to convert categorical values into a 1-dimensional numerical vector. The resulting vector will have only one element equal to 1 and the rest will be 0.

```python
# Here is a list of categorical features we need to encode
X = ['Dog', 'Cat', 'Bird']

# After encoding each feature will be represented as a number
Dog = [1, 0, 0]

Cat = [0, 1, 0]

Bird = [0, 0, 1]
```

### One-hot Encoding in Pandas

```python
import pandas as pd
import numpy as np

A = np.linspace(2.0, 10.0, num=3)
B = ['dog','cat','bird']
d = {'numeric': A, 'categorical': B}

df = pd.DataFrame(d)
df.head()
```

We apply one-hot encoding with the pandas get_dummies() method:

```python
print("Original features:\n", list(df.columns), "\n")
data_dummies = pd.get_dummies(df)
print("Features after get_dummies:\n", list(data_dummies.columns))
```

You can see that the continuous numeric features not touched, while the categorical features were expanded into one  new feature for each possible value:

```python
data_dummies.head()
```

# Scaling data

When we looked at neural networks and SVMs the data had to be scaled for the algorithms to be efficient. The same applies to dimensionality reduction and PCA.

`Standard Scaler`: Scales each data point to have a mean value 0 and standard deviation of 1.

`Min Max Scaler`: Scales each data point to a value between 0 and 1.

`Robust Scaler`: Similar to Standard Scaler but uses the median and quartiles, instead of mean and variance. This  makes the RobustScaler ignore data points that are very different from the rest.

`Normalizer`: Scales each datapoint such that the feature vector has a Euclidean length of 1. In other words,  it projects a data point on the circle.

# Principal Component Analysis (PCA)

Transforming and reducing the dimensionality is an important part of `Unsupervised Learning`. The most common motivations are visualization, compressing the data, and finding a representation that is more informative for further processing.

Principal Component Analysis (PCA) is one of the most widely used algorithms to achieve this.

Principal  component  analysis  is  a  method  that  rotates  the  dataset  in  a  way  such  that the  rotated  features  are  statistically  uncorrelated.  This  rotation  is  often  followed  by selecting  only  a  subset  of  the  new  features,  according  to  how  important  they  are  for explaining  the  data.

Let's say with have large dataset with alot of features, the idea of PCA is that we can somehow select the 'most important' directions, and keep those, while throwing away the ones that contribute mostly noise. 

### How PCA works
From a high-level view PCA has three main steps:

(1) Compute the covariance matrix of the data

(2) Compute the eigen values and vectors of this covariance matrix

(3) Use the eigen values and vectors to select only the most important feature vectors and then transform your data onto those vectors for reduced dimensionality!

Great video for understanding the math behind PCA: https://www.youtube.com/watch?v=FgakZw6K1QQ

PCA tries to maximize variance between principal components:
![image.gif](https://i.stack.imgur.com/Q7HIP.gif)

Let's use PCA on the breast cancer dataset:

```python
from sklearn.datasets import load_breast_cancer

cancer = load_breast_cancer()
cancer_df = pd.DataFrame(np.c_[cancer['data'], cancer['target']],
                  columns= np.append(cancer['feature_names'], ['target']))

cancer_df.describe()
```

Before we apply PCA, we scale our data so that each feature has unit variance using `StandardScaler`:

```python
from sklearn.preprocessing import StandardScaler

# Scale data
scaler = StandardScaler()
scaler.fit(cancer.data)
X_scaled = scaler.transform(cancer.data)

# Let's convert the scaled data to a pandas dataframe so we can visualize it
scaled_cancer_df = pd.DataFrame(np.c_[X_scaled, cancer['target']],
                  columns= np.append(cancer['feature_names'], ['target']))

display(scaled_cancer_df.describe())
```

The data now has a mean value 0 and standard deviation of 1.

Let's look at the first two features of the cancer data:

```python
import matplotlib.pyplot as plt
import mglearn

plt.figure(figsize=(8, 8))
mglearn.discrete_scatter(X_scaled[:, 0], X_scaled[:, 1], cancer.target)
plt.legend(cancer.target_names, loc="best")
plt.gca().set_aspect("equal")
plt.xlabel("First principal component")
plt.ylabel("Second principal component")
```

As you can see the data isn't split very well, we can improve this with PCA:

```python
from sklearn.decomposition import PCA

# keep the first two principal components of the data
pca = PCA(n_components=2)
# fit PCA model to breast cancer data
pca.fit(X_scaled)
# transform data onto the first two principal components
X_pca = pca.transform(X_scaled)
print("Original shape: {}".format(str(X_scaled.shape)))
print("Reduced shape: {}".format(str(X_pca.shape)))

print("\nPCA sorts the principal components based on variance (how much the data is spread out),", 
      "\nso it keeps the features with the most information:")
print(pca.explained_variance_ratio_)  
```

```python
import matplotlib.pyplot as plt
import mglearn

plt.figure(figsize=(8, 8))
mglearn.discrete_scatter(X_pca[:, 0], X_pca[:, 1], cancer.target)
plt.legend(cancer.target_names, loc="best")
plt.gca().set_aspect("equal")
plt.xlabel("First principal component")
plt.ylabel("Second principal component")
```

# Clustering

## K-means clustering

k-means clustering is one of the simplest and most commonly used clustering algorithms. It tries to find cluster  centers that are representative of certain region of the data.  

The algorithm alternates between two steps: assigning each data point to the closest cluster center, and then setting  each cluster center as the mean of the datapoints that are assigned to it. 

![image.gif](https://zippy.gfycat.com/InbornCloseFlickertailsquirrel.gif)

```python
import mglearn
mglearn.plots.plot_kmeans_algorithm()
```

### K-means algorithm:

1. Initialize n points (clusters) with random centers.
2. Repeat until converged (the cluster centers remain unchanged)
    1. Assign each data point to te center point it is closest to.
    2. Update the cluster centers to the mean of their assigned points.

### K-means in sklearn

Let's first create some random data with 2 features:

```python
from sklearn.datasets import make_blobs 
from sklearn.cluster import KMeans
# generate synthetic two-dimensional data
X, y = make_blobs(n_samples=200, centers=5,cluster_std=1.1, n_features=2, random_state=1)

plt.figure(figsize=(8, 8))
mglearn.discrete_scatter(X[:, 0], X[:, 1])
```

```python
# build the clustering model
kmeans = KMeans(n_clusters=3) 
kmeans.fit(X)
# Predict the clusters the data points belong to
y_kmeans = kmeans.predict(X)

# Plot the clusters
plt.figure(figsize=(8, 8))
plt.scatter(X[:, 0],X[:, 1], c=y_kmeans, s=50, cmap='viridis')

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5);
```

### Determining The Optimal Number Of Clusters

#### Elbow method


1. Compute clustering algorithm (e.g., k-means clustering) for different values of k. For instance, by varying k from 1 to 10 clusters.
2. For each k, calculate the total within-cluster sum of square (wss).
3. Plot the curve of wss according to the number of clusters k.
4. The location of a bend (knee) in the plot is generally considered as an indicator of the appropriate number of clusters.



```python
Sum_of_squared_distances = []
K = range(1,15)
for k in K:
    km = KMeans(n_clusters=k)
    km = km.fit(X)
    Sum_of_squared_distances.append(km.inertia_)
```

```python
plt.plot(K, Sum_of_squared_distances, 'bx-')
plt.xlabel('k')
plt.ylabel('Sum of squared distances')
plt.title('Elbow Method For Optimal k')
plt.show()
```

The numbers at the lower end og the elbow is 4 and 5, so let's create a K-Means model with 5 clusters.

```python
# build the clustering model
kmeans = KMeans(n_clusters=5) 
kmeans.fit(X)
# Predict the clusters the data points belong to
y_kmeans = kmeans.predict(X)

# Plot the clusters
plt.figure(figsize=(8, 8))
plt.scatter(X[:, 0],X[:, 1], c=y_kmeans, s=50, cmap='viridis')

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5);
```

## Guassian Mixing

The `K-Means` clustering model  is simple and relatively easy to understand, but its simplicity leads to practical challenges in its application. 

In particular, the non-probabilistic nature of k-means and its use of simple distance-from-cluster-center to assign cluster membership leads to poor performance for many real-world situations.

`Gaussian Mixing` can be viewed as an extention to k-means.

Let's look at a dataset where k-means will fail.

```python
import numpy as np

X, y = make_blobs(n_samples=200, random_state=170) # cluster_std=2 for gaussian probability example

# Anisotropicly distributed data
transformation = [[0.8, -0.6], [-0.4, 0.8]]
X_streched = np.dot(X, transformation)

plt.figure(figsize=(8, 8))
mglearn.discrete_scatter(X_streched[:, 0], X_streched[:, 1], y)
```

```python
# build the clustering model
kmeans = KMeans(n_clusters=3)
kmeans.fit(X_streched)
# Predict the clusters the data points belong to
y_kmeans = kmeans.predict(X_streched)

# Plot the clusters
plt.figure(figsize=(8, 8))
plt.scatter(X_streched[:, 0],X_streched[:, 1], c=y_kmeans, s=50, cmap='viridis')

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=400, alpha=0.5);
```

As you can see `K-means` performs pretty poorly if the data is streched like this becuase the clusters are defined in a circle around the centers.

A `Gaussian mixture` model (GMM) attempts to find a mixture of multi-dimensional Gaussian probability distributions that best model any input dataset.

### Gaussian mixture algorithm:

1. Choose starting guesses for the location and shape

1. Repeat until converged:
    1. E-step: for each point, find weights encoding the probability of membership in each cluster
    2. M-step: for each cluster, update its location, normalization, and shape based on all data points, making use of the weights


Let's try with he `Gaussian mixture` model with sklearn:

```python
from sklearn.mixture import GaussianMixture
gmm = GaussianMixture(n_components=3, covariance_type='full')
gmm.fit(X_streched)
labels = gmm.predict(X_streched)

plt.figure(figsize=(8, 8))
plt.scatter(X_streched[:, 0], X_streched[:, 1], c=labels);
```

```python
import matplotlib.pyplot as plt
import PlotGMM

plt.figure(figsize=(8, 8))
PlotGMM.plot_gmm(gmm, X_streched)
```

Because GMM is a probabilistic model, it is also possible to see the probabilites of the data points belonging to a cluster:

```python
probs = gmm.predict_proba(X_streched)
print(probs[:5].round(3))
```

### Covariance type

The covariance type controls the degrees of freedom in the shape of each cluster; it is essential to set this carefully for any given problem.

`covariance_type='spherical'`: the contour is a sphere.

`covariance_type='diag'`: the contour is aligned to the axes. Means that the size of the cluster along each dimension can be set independently

`covariance_type='tied'`: all components share the same general covariance matrix. Every cluster has the same ellipse.

`covariance_type='full'`: each component has its own general covariance matrix. Allows each cluster to be modeled as an ellipse with arbitrary orientation.

![covariances](https://scikit-learn.org/stable/_images/sphx_glr_plot_gmm_covariances_001.png)

### Determining the optimal number of clusters

A good way to detemining the optimal number of clusters for Gaussian Mixing is to do cross validation and adjust the model likelihoods using some analytic criterion such as the __[Akaike information criterion (AIC)](https://en.wikipedia.org/wiki/Akaike_information_criterion)__ or the __[Bayesian information criterion (BIC)](https://en.wikipedia.org/wiki/Bayesian_information_criterion)__. 

Scikit-Learn's GMM estimator actually includes built-in methods that compute both of these, and so it is very easy to operate on this approach.

```python
import numpy as np

X, y = make_blobs(centers=8, n_samples=200, cluster_std=1.4, random_state=9) # cluster_std=2 for gaussian probability example

plt.figure(figsize=(8, 8))
mglearn.discrete_scatter(X[:, 0], X[:, 1])
```

```python
n_components = np.arange(1, 13)
models = [GaussianMixture(n, covariance_type='full', random_state=0).fit(X)
          for n in n_components]

plt.plot(n_components, [m.bic(X) for m in models], label='BIC')
plt.plot(n_components, [m.aic(X) for m in models], label='AIC')
plt.legend(loc='best')
plt.xlabel('n_components');
```

The optimal number of clusters is the value that minimizes the AIC or BIC, depending on which approximation we wish to use. In this case we can see that a model with between 5 and 7 components is best.

Notice the important point: this choice of number of components measures how well GMM works as a density estimator, not how well it works as a clustering algorithm. 

```python
gmm = GaussianMixture(5, covariance_type='full', random_state=0)
gmm.fit(X)
labels = gmm.predict(X)

plt.figure(figsize=(8, 8))
plt.scatter(X[:, 0], X[:, 1], c=labels);
```

## Gaussian Mixture models for generating new data

The gaussian mixture model can be used to generate new data by using:
`gmm.sample(num_samples)` on  a trained GMM model:

```python
generated_X, generated_y = gmm.sample(10000)

plt.figure(figsize=(8, 8))
plt.scatter(generated_X[:, 0], generated_X[:, 1], c=generated_y);
```

Let's use this on a more interesting example, where we try to generate new handwritten digits after training a GMM model on the digits dataset:

```python
from sklearn.datasets import load_digits
digits = load_digits()
digits.data.shape
```

```python
fig, ax = plt.subplots(2, 3, figsize=(4, 4))
for i, axi in enumerate(ax.flat):
    im = axi.imshow(digits.data[i].reshape(8, 8), cmap='binary')
```

The data has 64 dimensions (8x8 pixels).
Gaussian mixture might struggle converging in such a high dimensional space, so let's use PCA to reduce the dimensions. (Gaussian Mixture works fine without PCA with this dataset, so this is more as an example)

```python
from sklearn.decomposition import PCA
pca = PCA(0.98, whiten=True) # variance that needs to be explained is greater than 0.98 (removes features that doesn't have much information)
data = pca.fit_transform(digits.data)
data.shape
```

```python
# Fit handwritten images to model
gmm = GaussianMixture(200, covariance_type='full', random_state=0)
gmm.fit(data)
```

Now let's create some new digits by sampling the our trained model:

```python
new_digits, new_y = gmm.sample(16)
# For the new digits to make any sense we need to perform an inverse transform with the pca
new_digits = pca.inverse_transform(new_digits)
```

```python
fig, ax = plt.subplots(2, 8, figsize=(10, 3))
for i, axi in enumerate(ax.flat):
    im = axi.imshow(new_digits[i].reshape(8, 8), cmap='binary')
```
